```python
import vega_datasets
import pandas as pd

bs = vega_datasets.data.birdstrikes()

# Mengecek apakah data ada yang not available
bs.isna()

# Jika ada data yg Not Available maka di ganti menjadi nilai data rata-rata
bs['Speed_IAS_in_knots'] = bs['Speed_IAS_in_knots'].fillna(bs['Speed_IAS_in_knots'].mean())

# Menampilkan seluruh data ke table
pd.DataFrame(bs)
```
